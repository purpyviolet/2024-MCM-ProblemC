```python
import pandas as pd
import matplotlib.pyplot as plt

# 从CSV文件中读取数据
file_path = 'men_shap.csv'
df = pd.read_csv(file_path)

# 获取相关性列
feature_importance = df.set_index('classify')['shap_values'].to_dict()

# 对特征重要性进行排序
sorted_features = sorted(feature_importance.items(), key=lambda x: x[1], reverse=True)

# 选择前几个特征和设定一个阈值(修改这里来控制显示的多少)
threshold = 0.1
top_features = {feature: value for feature, value in sorted_features if value > threshold}

# 对结果进行反转
sorted_top_features = dict(reversed(list(top_features.items())))

# 绘制横向条形图
plt.figure(figsize=(10, 6))
bars = plt.barh(list(sorted_top_features.keys()), list(sorted_top_features.values()), color='skyblue')

# 设置每个条形的数值标签
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, round(bar.get_width(), 3), ha='left', va='center', color='black')

# 设置图表标题和坐标轴标签
plt.title('The correlation between features and fluctuations(men)')
plt.xlabel('Shap Value')
plt.ylabel('Features')

# 隐藏坐标轴上的"Others"标签
plt.yticks([key for key in sorted_top_features.keys() if key != 'Others'])

# 显示图表
plt.show()

```
